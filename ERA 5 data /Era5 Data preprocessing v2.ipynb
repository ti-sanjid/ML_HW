```python
import os
from tqdm.auto import tqdm

# # Define the directory path
directory_path = '/mnt/1A42C1DD42C1BE2F/MyProjects/ML_HEATWAVE/ERA 5 /Geopotential/'

# Loop through all files in the directory
for filename in tqdm(os.listdir(directory_path)):
    if filename.endswith('.nc'):  # Assuming you want to process only netcdf files
        file_path = os.path.join(directory_path, filename)
        # Process the file here
        ds = xr.open_dataset(file_path)
        ds_daily = ds.z.resample(time='D').mean() #change 
        ds_daily = ds_daily.sel(time=(ds_daily['time.month'] >= 3) & (ds_daily['time.month'] <= 10))
        

        new_lat = np.arange(26,20.26, -1.25).astype(np.float32)
        new_lon = np.arange(88.50,92.76, 1.25).astype(np.float32)
        resized_ds = ds_daily.interp(latitude=new_lat, longitude=new_lon)
        dfs=[]
        for level in [1000, 850, 500]:
            #change
            data_level = resized_ds.sel(level=level).to_dataframe()

            data_level = data_level.drop('level', axis=1)
            #change
            data_level = data_level.rename(columns={'z': f'geo_hgt_{level}hpa'})
            data_level = data_level.reset_index()
            dfs.append(data_level)

        merged_df = dfs[0]
        for df in dfs[1:]:
            merged_df = pd.merge(merged_df, df)
        # change
        merged_df.to_csv(f'/mnt/1A42C1DD42C1BE2F/MyProjects/ML_HEATWAVE/ERA 5 /Geopotential/csv files/{filename[:-3]}.csv',index=False)
  
    
    
    
    
```

```python
def rhum_2m_calculation(t2m,d2m):
    
    t2m_in_celcelsius=t2m-273.15
    d2m_in_celcelsius=d2m-273.15
    exponent_svp=(7.5*t2m_in_celcelsius)/(237.3+t2m_in_celcelsius)
    exponent_avp=(7.5*d2m_in_celcelsius)/(237.3+d2m_in_celcelsius)
    svp=6.11*10**exponent_svp
    avp=6.11*10**exponent_avp
    rhum_2m=(avp/svp)*100
    return rhum_2m
    
    
# heat_index(315,305)    
    
```

```python
def heat_index(t2m_kelvin, rh_percentage):
    
    t_fahrenheit = (t2m_kelvin-273.15) * 9/5 + 32 
    
    # Rothfusz regression equation
    heat_index = -42.379 + 2.04901523*t_fahrenheit + 10.14333127*rh_percentage - \
                 0.22475541*t_fahrenheit*rh_percentage - 0.00683783*t_fahrenheit**2 - \
                 0.05481717*rh_percentage**2 + 0.00122874*t_fahrenheit**2*rh_percentage + \
                 0.00085282*t_fahrenheit*rh_percentage**2 - 0.00000199*t_fahrenheit**2*rh_percentage**2
    
    # Apply adjustments
    if rh_percentage < 13 and 80 <= t_fahrenheit <= 112:
        adjustment = ((13 - rh_percentage) / 4) * math.sqrt((17 - abs(t_fahrenheit - 95)) / 17)
        heat_index -= adjustment
    
    if rh_percentage > 85 and 80 <= t_fahrenheit <= 87:
        adjustment = ((rh_percentage - 85) / 10) * ((87 - t_fahrenheit) / 5)
        heat_index += adjustment
    heat_index=(heat_index-32)*5/9+273.15
    
    return heat_index

```

```python
import math

def heat_index(T, RH):
    T = (T-273.15) * 9/5 + 32 # converting calvin to farenheit.
    
    # Rothfusz regression equation
    HI = -42.379 + 2.04901523*T + 10.14333127*RH - 0.22475541*T*RH - 0.00683783*T**2 - \
         0.05481717*RH**2 + 0.00122874*T**2*RH + 0.00085282*T*RH**2 - 0.00000199*T**2*RH**2
    
    # Adjustments
    if RH < 13 and 80 <= T <= 112:
        ADJUSTMENT = ((13 - RH) / 4) * math.sqrt((17 - abs(T - 95)) / 17)
        HI -= ADJUSTMENT
    
    if RH > 85 and 80 <= T <= 87:
        ADJUSTMENT = ((RH - 85) / 10) * ((87 - T) / 5)
        HI += ADJUSTMENT
    
    # Check if Rothfusz regression is not appropriate
    if HI < 80:
        # Apply simpler formula
        HI = 0.5 * (T + 61.0 + ((T - 68.0) * 1.2) + (RH * 0.094))
    
    return HI

```
