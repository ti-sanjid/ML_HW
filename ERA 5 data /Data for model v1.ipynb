# HeatWave Detection - 1

```python
from tqdm.auto import tqdm

duration = 0
index_list = []
HW_dates=[]

for lat in tqdm(lat_values, desc='Latitudes',leave=True):
    for lon in tqdm(lon_values, desc='Longitudes',leave=False):
        for date in tqdm(dates, desc='Dates', leave=False):
            date_mask = df['time'] == date
            lat_mask = df['latitude'] == lat
            lon_mask = df['longitude'] == lon
            
            if (df[date_mask & lat_mask & lon_mask]['mx2t'].values >= 307.054055).any() and (df[date_mask & lat_mask & lon_mask]['mn2t'].values >= 300.864264).any():
                duration += 1
                index_list.extend(df[date_mask & lat_mask & lon_mask].index)
#                 if duration >=3:
#                     df.loc[index_list,'Heatwave']=1
            else:               

                if duration>=3:
                    df.loc[index_list,'Heatwave']=1
                    HW_dates.append(df['time'][index_list[0]])
                    print('Duration:',duration,'days |','Starting date:',df['time'][index_list[0]],'| lat:',df['latitude'][index_list[0]],'lon:',df['longitude'][index_list[0]])

                    date_mask=df['time'].isin(df['time'][index_list[:]])
                    df_check=df[date_mask]
                    
                    if (df_check['mx2t'].values>=307.054055).any() or (df_check['mn2t'].values>=300.864264).any() :
                    
                        condition = (df_check['mx2t'] >=307.054055) | (df_check['mn2t'] >= 300.864264)
                        indices = df_check.index[condition]
                        df.loc[indices,'Heatwave']=1
                        latitudes=df_check.loc[indices,'latitude'].values
                        longitudes=df_check.loc[indices,'longitude'].values
                        for i, j in zip(latitudes, longitudes):
                            print('Surrounding Locations:')
                            print('Latitude:',i, '| Longitude:', j)


                duration = 0
                index_list = []                    
                        
```
