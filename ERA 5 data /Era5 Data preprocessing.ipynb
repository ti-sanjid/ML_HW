```python
import os
from tqdm.auto import tqdm

# # Define the directory path
directory_path = '/mnt/1A42C1DD42C1BE2F/MyProjects/ML_HEATWAVE/ERA 5 /Era5 Temperature'
#change

# Loop through all files in the directory
for filename in tqdm(os.listdir(directory_path)):
    if filename.endswith('.nc'):  # Assuming you want to process only netcdf files
        file_path = os.path.join(directory_path, filename)
        # Process the file here
        ds = xr.open_dataset(file_path)
        ds_daily = ds.t2m.resample(time='D').mean() #change 
        lat_range = slice(27.5,20.0)  # take lat values in backwards directions.
        lon_range = slice(87.5, 92.5)
        data_bangladesh = ds_daily.sel(latitude=lat_range,longitude=lon_range)
        new_lat = np.arange(27.5,19.9, -2.5).astype(np.float32)
        new_lon = np.arange(87.5, 92.6, 2.5).astype(np.float32)
        resized_ds = data_bangladesh.interp(latitude=new_lat, longitude=new_lon)
        df = resized_ds.to_dataframe()
        df=df.reset_index()
        #change
        df.to_csv(f'/mnt/1A42C1DD42C1BE2F/MyProjects/ML_HEATWAVE/ERA 5 /Era5 Temperature/t2m csv files/t2m.{filename[12:-3]}.csv')

        
    
    
    
    
```

```python
import os
from tqdm.auto import tqdm

# # Define the directory path
directory_path = '/mnt/1A42C1DD42C1BE2F/MyProjects/ML_HEATWAVE/ERA 5 /Geopotential/'

# Loop through all files in the directory
for filename in tqdm(os.listdir(directory_path)):
    if filename.endswith('.nc'):  # Assuming you want to process only netcdf files
        file_path = os.path.join(directory_path, filename)
        # Process the file here
        ds = xr.open_dataset(file_path)
        ds_daily = ds.z.resample(time='D').mean() #change 
        lat_range = slice(27.5,20.0)  # take lat values in backwards directions.
        lon_range = slice(87.5, 92.5)
        data_bangladesh = ds_daily.sel(latitude=lat_range,longitude=lon_range)
        new_lat = np.arange(27.5,19.9, -2.5).astype(np.float32)
        new_lon = np.arange(87.5, 92.6, 2.5).astype(np.float32)
        resized_ds = data_bangladesh.interp(latitude=new_lat, longitude=new_lon)
        dfs=[]
        for level in [1000, 850, 500]:
            #change
            data_level = resized_ds.sel(level=level).to_dataframe()

            data_level = data_level.drop('level', axis=1)
            #change
            data_level = data_level.rename(columns={'z': f'geo_hgt_{level}hpa'})
            data_level = data_level.reset_index()
            dfs.append(data_level)

        merged_df = dfs[0]
        for df in dfs[1:]:
            merged_df = pd.merge(merged_df, df)
        # change
        merged_df.to_csv(f'/mnt/1A42C1DD42C1BE2F/MyProjects/ML_HEATWAVE/ERA 5 /Geopotential/csv files/{filename[:-3]}.csv')
  
    
    
    
    
```
